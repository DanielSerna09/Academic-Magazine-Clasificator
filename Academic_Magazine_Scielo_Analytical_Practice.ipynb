```
#Aca se conecta a la HTML de la plataforma, y se busca la seccion donde se encuentra la informacion a extaer.
uClient=uReq(my_URL)
page_htlm = uClient.read()
uClient.close()
page_soup= soup(page_htlm,"html.parser")
page_soup.h
page_soup.section.div
Articulos = page_soup.find_all("div",{"class":"item"})
len (Articulos)
Container=Articulos[0]
```

```
#Aca lo que se hace es encontrar las informacion que se necesita,  como el titulo del articulo, el nombre de los autores, el abstrac etc. 
# Todo esto se lleva a un ciclo para que recorra en la pagina la cantidad de articulos a extraer.
# Es importante mencionar que la pagina Scielo solo deja extraer tandas de 100 articulos en cada corrida
#Un elemento importante a extraer es el Link y el DOI de cada uno de los articulos, dado que seran la llave para ingresar a cada uno de las urls y realizar la extraccion  de las referencias

# se comenta el codigo para guardar la informacion en un archivo.csv

#filename="Articulos 1-100-2014.csv"
#f=open (filename,"w",encoding="utf-8")
#headers="TitleID,Year,DOI,Journal,Len_Abstract,Link,LenAuthors,Title\n"
#f.write(headers)


for Container in Articulos:
    #Container.find("a",{"class":"author"})
    TitleID=Container.strong["id"]
    Title=Container.strong.text.strip()
    date=Container.find_all("span",{"style":"margin: 0"})
    #date[0].text.strip()
    Year=date[1].text.strip()
    DOI=Container.find("span",{"class":"DOIResults"})
    if DOI is None:
        DOI='N'
    else:
        DOI=Container.find("span",{"class":"DOIResults"}).text.strip()
    Journal=Container.find_all("",{"class":"showTooltip dropdown-toggle"})
    Journal2=Journal[2].text.strip()
    Abstract=Container.find("div",{"class":"abstract"}).text.strip()
    LenAbstract=len(Abstract)
    Authors=Container.find("div",{"class":"line authors"})
    if Authors is None:
        Authors='N'
    Link=Container.a["href"]
    LenAuthors=len(Authors)
    #print("Authors: "+Authors)
    print("TitleID: "+TitleID)
    print("Title: "+Title)
    print("Year: "+Year)
    print("DOI :",DOI)
    print("Journal: "+Journal2)
    print("LenAbstract: ",LenAbstract)
    print("Link: "+Link)
    #print("Abstract: "+Abstract)
    print("LenAuthors: ",LenAuthors)
#    f.write(TitleID + ","+ Year.replace(",","") + "," + DOI + "," + Journal2 + "," + str(LenAbstract) + "," + Link + "," + str(LenAuthors) + "," + Title +"\n")
#f.close
```
